```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.finance import candlestick_ohlc
import pandas as pd
import numpy as np
from time import mktime
from datetime import date, datetime
from sklearn.externals import joblib
from poloniex import Poloniex
import os
import time
# %matplotlib inline
polo_api = os.environ['POLO_API']
polo_secret = os.environ['POLO_SECRET_KEY']
polo = Poloniex(key = polo_api, secret = polo_secret)
start = mktime(date(2015,8,8).timetuple())
end = mktime(date(2017,6,10).timetuple())
df = pd.DataFrame.from_records(polo.returnChartData(currencyPair = 'USDT_BTC',start = start, end = end, period = 86400), index = 'date')
df['date']= df.index
df = df.apply(lambda s: pd.to_numeric(s))
# df.date = df.date.apply(lambda x: datetime.utcfromtimestamp(x))
df = df[['date','open','high','low','close']]
```

```python
ohlc = df.values
fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(111)
```

```python
candlestick_ohlc(ax, ohlc,width=0.4, colorup='#77d879', colordown='#db3f3f', alpha = 0.8)
plt.show()
```

```python
from math import pi
from bokeh.plotting import figure, show, output_file

inc = df.close > df.open
dec = df.open > df.close

w = 14400000
TOOLS = 'pan, wheel_zoom, box_zoom, reset, save'
p = figure(x_axis_type = 'datetime', tools = TOOLS, plot_width = 1000, title = 'BTC Candlestick')
p.xaxis.major_label_orientation = pi/4
p.grid.grid_line_alpha = 0.3

p.segment(df.date,df.high,df.date, df.low, color = 'black')
p.vbar(df.date[inc], w, df.open[inc], df.close[inc], fill_color = 'green', line_color = 'black')
p.vbar(df.date[dec], w, df.open[dec], df.close[dec], fill_color = '#f2583E', line_color = 'black')
output_file('BTCcandlestick.html', title = 'BTC Candlestick')
```
